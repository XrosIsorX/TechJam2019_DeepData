# Group

```python
def discretize(row, step_list, column):
    for i, step in enumerate(step_list):
        if row[column] < step:
            return i
    return len(step_list)
```

```python
step_list = [i for i in range(5,100, 5)] + [i for i in range(1000,12000, 1000)]
column = "kp_txn_mean"

train_df["kp_txn_mean_group"] = train_df.apply(discretize, args=[step_list, column], axis=1)
test_df["kp_txn_mean_group"] = test_df.apply(discretize, args=[step_list, column], axis=1)

step_list = [i for i in range(3,50, 3)]
column = "cc_txn_cnt"

train_df["cc_txn_cnt_group"] = train_df.apply(discretize, args=[step_list, column], axis=1)
test_df["cc_txn_cnt_group"] = test_df.apply(discretize, args=[step_list, column], axis=1)
train_df

step_list = [i for i in range(500,8000, 500)]
column = "cc_txn_mean"

train_df["cc_txn_mean_group"] = train_df.apply(discretize, args=[step_list, column], axis=1)
test_df["cc_txn_mean_group"] = test_df.apply(discretize, args=[step_list, column], axis=1)
train_df

step_list = [i for i in range(10,140, 10)]
column = "kp_txn_count"

train_df["kp_txn_count_group"] = train_df.apply(discretize, args=[step_list, column], axis=1)
test_df["kp_txn_count_group"] = test_df.apply(discretize, args=[step_list, column], axis=1)
train_df
```

# Find all statistic

```python
def cal_all_stats(df, feature, feature_value, dic):
    temp_df = df[df[feature] == feature_value]
    dic[f"{feature}_mode"][feature_value] = temp_df[config.target].mode()
    dic[f"{feature}_mean"][feature_value] = temp_df[config.target].mean()
    dic[f"{feature}_per25"][feature_value] = temp_df[config.target].quantile(0.25)
    dic[f"{feature}_per35"][feature_value] = temp_df[config.target].quantile(0.35)
    dic[f"{feature}_per40"][feature_value] = temp_df[config.target].quantile(0.4)
    dic[f"{feature}_per45"][feature_value] = temp_df[config.target].quantile(0.45)
    dic[f"{feature}_per50"][feature_value] = temp_df[config.target].quantile(0.5)
    dic[f"{feature}_per60"][feature_value] = temp_df[config.target].quantile(0.6)
    dic[f"{feature}_per75"][feature_value] = temp_df[config.target].quantile(0.75)
```

```python
stat_list = [    "mode", 
                 "mean", 
                 "per25", 
                 "per35", 
                 "per40", 
                 "per45", 
                 "per50", 
                 "per60", 
                 "per75",
            ]

stat_dict = {}
feature_list = ["age", "cc_count"]
for feature in feature_list:
    for stat in stat_list:
        stat_dict[feature + "_" + stat] = {}

for feature in feature_list:
    for age in age_list:
        cal_all_stats(train_df, feature, age, stat_dict)
```
