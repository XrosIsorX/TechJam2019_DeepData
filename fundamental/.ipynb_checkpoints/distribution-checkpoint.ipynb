```python
import pandas as pd
import matplotlib.pyplot as plt

import config
import utils.DataMaker as dm
```

```python
train_income_df, test_income_df, demographic_df, cc_df, kplus_df, train_df, test_df = dm.read_all()
```

```python
age_list = train_df["age"].unique()
age_list = sorted(age_list)
```

# Find all statistic

```python
stats = {
    "age_mode": {},
    "age_mean": {},
    "age_per25": {},
    "age_per35": {},
    "age_per40": {},
    "age_per45": {},
    "age_per50": {},
    "age_per60": {},
    "age_per75": {},
}
for age in age_list:
    temp_df = train_df[train_df["age"] == age]
    stats["age_mode"][age] = temp_df["income"].mode()
    stats["age_mean"][age] = temp_df["income"].mean()
    stats["age_per25"][age] = temp_df["income"].quantile(0.25)
    stats["age_per35"][age] = temp_df["income"].quantile(0.35)
    stats["age_per40"][age] = temp_df["income"].quantile(0.4)
    stats["age_per45"][age] = temp_df["income"].quantile(0.45)
    stats["age_per50"][age] = temp_df["income"].quantile(0.5)
    stats["age_per60"][age] = temp_df["income"].quantile(0.6)
    stats["age_per75"][age] = temp_df["income"].quantile(0.75)
```

# Mean

```python
test_df["income"] = test_df.apply(lambda row : int(stats["age_mean"][row["age"]]), axis=1)
test_df[["income"]].to_csv("age_mean.csv")
test_df
```

# Mode

```python
test_df["income"] = test_df.apply(lambda row : int(stats["age_mode"][row["age"]]), axis=1)
test_df[["income"]].to_csv("age_mode.csv")
test_df
```
