# Mean by feature list

```python
mean_dict = {}
quan45_dict = {}
feature_list = ["age", "ocp_cd"]
for feature in feature_list:
    temp_df = train_df[[config.target, feature]].groupby(feature).mean()
    mean_dict[feature] = temp_df
    
    temp_df = train_df[[config.target, feature]].groupby(feature).quantile(0.45)
    print(temp_df)
    quan45_dict[feature] = temp_df
```

```python
group_mean_df = train_df.groupby(feature_list).quantile(0.45)["income"]
group_mean_df
```

```python
def get_income(row):
    key = []
    for feature in feature_list:
        key.append(row[feature])
    key = tuple(key)
    if key in group_mean_df:
        return int(group_mean_df[key])
    else:
        print(key)
        return int(quan45_dict["age"][config.target][row["age"]])
```
