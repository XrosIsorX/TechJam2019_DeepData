```python
import numpy as np
from sklearn.linear_model import LinearRegression

import pandas as pd
import matplotlib.pyplot as plt

import config
import utils.DataMaker as dm
```

```python
train_income_df, test_income_df, demographic_df, cc_df, kplus_df, train_df, test_df = dm.read_all()
```

```python
train_df
```

```python
ocp_cd_list = sorted(train_df["ocp_cd"].unique())
ocp_cd_list
```

```python
model_dict = {}
for ocp_cd in ocp_cd_list:
    temp_df = train_df[train_df["ocp_cd"] == ocp_cd]
    linear_model = LinearRegression(normalize=True)
    linear_model.fit(temp_df[["age", "kp_txn_mean"]], temp_df["income"])
    model_dict[ocp_cd] = linear_model
```

```python
def predict_linear(row, model_dict):
    return int(model_dict[row["ocp_cd"]].predict([row[["age", "kp_txn_mean"]]])[0])
```

```python
test_df["income"] = test_df.apply(predict_linear, args=[model_dict], axis=1)
```

```python
test_df[["income"]].to_csv("linear_norm_-group_ocp_cd-_-inp_age_kp_txn_mean-.csv")
```
